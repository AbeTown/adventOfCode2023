### gameplan
- think I will read it into a 2d array...
- initially thought looking for symbols first might be more clever but think just looking for numbers is smarter.
- so left to right, top to bottom, we will go through and find each number, and check for symbols arounnd it, and add it to total if found.
- when we find a number, save first digit, figure out how many digits are in the number. (how_many_digits(startIndexX, startIndexY))
- when we have start index, and number of digits, e.g. 3rd line, 3th character, 4 digits, => chars 3, 4, 5, 6, then we add +2 to length, and look for symbols at indicies on line before, after, and current
- so look at 2, 3, 4, 5, 6, 7 on lines 2, 3, 4. If we find any symbols (is_symbol(char) checks if char is !numeric and ! a period)
- we can pass start index and length to a helper as well (check_surroundings(startx, starty, length))
- how_many_digits should figure out the number too and pass it back... e.g. 238
- if we find a symbol, add our number to the running total.
- regardless, jump the "current index" forward so we don't count the same number again. alternatively, replace all of the numbers with periods. 

```python
#read into 2darray
engine = []
input = open('p3input.txt','r')
#print(input.read())
for line in input:
    engine.append(line)
    #remove new line chars
    #if(len(listline)>140):
    #    del listline[-1]
    #engine.append(listline)

#print(engine)
# check to see if every line is same length still...
#for line in engine:
#    print(line)

#check dimensions
print(len(engine))
print(len(engine[0]))
print(len(engine[139]))
print(range(len(engine)+2)) #0-141
print(range(len(engine[1])+2)) #0-142
dotarray = [["." for i in range(len(engine)+2)] for j in range(len(engine[1])+1)]
print(len(dotarray))
print(len(dotarray[0]))
print(len(dotarray[141]))
```

```python
def how_many_digits(x,y):
    line = engine[y]
    digits = 1
    if(not line[x].isnumeric()):
        print("ERRORRRRRR")

    number = line[x]
    stillNumeric = True
    while(stillNumeric):
        x = x+1
        if(not line[x].isnumeric()):
            return(digits, number)
        else:
            #increment digits and add latest digit to number
            digits += 1
            number = number + line[x]


print(how_many_digits(3,0))
            
```

```python
#x = 0
y = 0
total = 0
symbols = "@#/*%=$+-&+"

for line in engine:
    #print("line: " + line)
    x = 0
    while x < 140:
        #print("x is: " + str(x))
        char = line[x]
        #print("char is: " + char)
        if char.isnumeric():
            #print("char is numeric")
            #found number, time to do all the things
            digits = 0
            startIndex = [x, y]
            digitsResponse = how_many_digits(x, y)
            #print(digitsResponse)
            digits = digitsResponse[0]
            number = int(digitsResponse[1])
            addNumber = False
            topOfRangeY = 140 if y + 2 > 140 else y + 2
            bottomOfRangeY = 0 if y - 1 < 0 else y - 1
            #print("topy, boty: " + str(topOfRangeY) + ", " + str(bottomOfRangeY))
            for ycoord in range(bottomOfRangeY, topOfRangeY):
                    #print("ycoord: " + str(ycoord))
                    #we want to go 1 on either side of the digits 
                    topOfRangeX = 140 if x + digits + 1 > 140 else x + digits + 1
                    bottomOfRangeX = 0 if x - 1 < 0 else x -1
                    #print("topx, botx: " + str(topOfRangeX) + ", " + str(bottomOfRangeX))
                    for xcoord in range(bottomOfRangeX, topOfRangeX):
                        #print("ycoord, xcoord, and what is there: " + str(ycoord) + " " + str(xcoord) + " " + engine[ycoord][xcoord])
                        #this should be a box around the number
                        if(engine[ycoord][xcoord] in symbols):
                            #print("found symbol, adding")
                            addNumber = True
            if(addNumber):
                total += number
                print("total is: " + str(total))

            #either way, we want to skip the number of digits...
            #print("x is: " + str(x))
            x = x + digits - 1
            #print("x is: " + str(x))
        #about to go to next character, increment x counter
        x += 1
    #about to go to next line, increment y counter
    y += 1
```

## Puzzle 2
### gameplan
- go through and find the numbers like last time
- look for only * symbols
- when find, add to dictionary, where key is coords, and value is number that hit that gear
- at end, look for all gears with 2 numbers

```python
#x = 0
y = 0
total = 0
symbols = "*"
gearsDictionary = {} # (y,x) : [#1, #2, etc...]
for line in engine:
    #print("line: " + line)
    x = 0
    while x < 140:
        #print("x is: " + str(x))
        char = line[x]
        #print("char is: " + char)
        if char.isnumeric():
            digits = 0
            startIndex = [x, y]
            digitsResponse = how_many_digits(x, y)
            digits = digitsResponse[0]
            number = int(digitsResponse[1])
            topOfRangeY = 140 if y + 2 > 140 else y + 2
            bottomOfRangeY = 0 if y - 1 < 0 else y - 1
            #print("topy, boty: " + str(topOfRangeY) + ", " + str(bottomOfRangeY))
            for ycoord in range(bottomOfRangeY, topOfRangeY):
                    #print("ycoord: " + str(ycoord))
                    #we want to go 1 on either side of the digits 
                    topOfRangeX = 140 if x + digits + 1 > 140 else x + digits + 1
                    bottomOfRangeX = 0 if x - 1 < 0 else x -1
                    #print("topx, botx: " + str(topOfRangeX) + ", " + str(bottomOfRangeX))
                    for xcoord in range(bottomOfRangeX, topOfRangeX):
                        #print("ycoord, xcoord, and what is there: " + str(ycoord) + " " + str(xcoord) + " " + engine[ycoord][xcoord])
                        #this should be a box around the number
                        if(engine[ycoord][xcoord] in symbols):
                            #print("found symbol, adding")
                            if((ycoord,xcoord) in gearsDictionary):
                                gearsDictionary[(ycoord,xcoord)].append(number)
                            else:
                                gearsDictionary[(ycoord,xcoord)] = [number]

            #either way, we want to skip the number of digits...
            #print("x is: " + str(x))
            x = x + digits - 1
            #print("x is: " + str(x))
        #about to go to next character, increment x counter
        x += 1
        #print(gearsDictionary)
    #about to go to next line, increment y counter
    y += 1
print(gearsDictionary)
```
